```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = np.sqrt(X**2/4 + Y**2/9 + 1) 
Z1 = -np.sqrt(X**2/4 + Y**2/9 + 1)
print('Двуполостной гиперболоид')
ax.plot_surface(X,Y,Z)
ax.plot_surface(X,Y,Z1)
ax.scatter(0, 0, 0, 'z', 50, 'red')
show()
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
Y = np.arange(-5, 5, 0.5)
Z = np.arange(-5, 5, 0.5)
Z, Y = np.meshgrid(Z, Y)
X = np.sqrt(1 + Y**2/4) 
X1 = -np.sqrt(1 + Y**2/4) 
print('Гиперболический цилиндр')
ax.plot_surface(X,Y,Z)
ax.plot_surface(X1,Y,Z)
ax.scatter(0, 0, 0, 'z', 50, 'red')
show()
```
