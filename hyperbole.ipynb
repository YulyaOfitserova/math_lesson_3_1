```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
x = []
y = []
x2 = []
a = 2
b = 5
for i in range(-500, 500):
    y_ = i/100
    y.append(y_)
    x.append(math.sqrt((1+(y_**2/a**2)) * b**2))
    x2.append(-math.sqrt((1+(y_**2/a**2)) * b**2))
plt.plot(x,y)
plt.plot(x2,y)
plt.xlabel("x")
plt.ylabel("y")
```
