```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
x = []
y = []
y2 = []
a = 5
b = 3
for i in range(-500, 501):
    x_ = i/100
    x.append(x_)
    y.append(math.sqrt((1-(x_**2/a**2)) * b**2))
    y2.append(-math.sqrt((1-(x_**2/a**2)) * b**2))
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
```
