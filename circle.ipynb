```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
x = []
y = []
y2 = []
x0 = 0
y0 = 0
R = 5
for i in range(-500, 501):
    x_ = i/100
    x.append(x_)
    y.append(math.sqrt(R**2-(x_-x0)**2) + y0)
    y2.append(-math.sqrt(R**2-(x_-x0)**2) + y0)
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
```
